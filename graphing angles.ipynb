```python
f = open('raw_angles.txt', 'r')
content = f.read()
words = content.split()
numbers = []
for word in words:
    number = int(word)
    numbers.append(number)
f.close()
#plt.plot(angles)
```

```python
plt.plot(numbers,'go')
#this is the angles of least crowding chosen by a gradient descent. notable shifting of lines. 6 by 4? NOT fibonacci..
#8 by 5 is an expected solution obtained by artificial means, also 3 by 5, 13 by 
```

```python
i = 0
difs = [0]
while i < len(numbers)-1:
    difference = numbers[i+1] - numbers[i]
    if difference < 0:
        difference += 360
    difs.append(difference)
    i += 1
plt.plot(difs,'bo')
#seems to hang around 75,190->150,and 240->270. golden angle is 137 degrees, a notable unshown result.
```
